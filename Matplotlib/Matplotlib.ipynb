```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(x, y)
ax.set(title='Testing', xlabel='x-axis', ylabel='y-axis')
fig.savefig('fig.jpg')
```
